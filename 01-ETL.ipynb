### Tratamientos en la columna HH

convertimos los valores a numerico y los errores a nan

```python
# Convertir la columna a números, manejar valores no numéricos asignándoles NaN
df_hechos_victimas['HH'] = pd.to_numeric(df_hechos_victimas['HH'], errors='coerce')
```

obtenemos los valores que no son nan para ver como rellenar los valores si con la media o mediana y esto dependera al coeficiente de variacion

```python
# Eliminar filas con valores NaN
df_validos = df_hechos_victimas.dropna(subset=['HH'])
# Calcular la media y la desviación estándar de la columna
media = df_validos['HH'].mean()
desviacion_estandar = df_validos['HH'].std()

# Calcular el coeficiente de variación
coeficiente_variacion = (desviacion_estandar / media) * 100
coeficiente_variacion
```

Al ser un valor mayor al 50% se decide usar la mediana por la dispercion de los datos

```python
# Calcular la mediana de la columna
mediana = int(df_hechos_victimas['HH'].median())

# Rellenar NaN con la mediana
df_hechos_victimas['HH'] = df_hechos_victimas['HH'].fillna(mediana)


# Convertir la columna 'HH' a tipo de dato entero
df_hechos_victimas['HH'] = df_hechos_victimas['HH'].astype(int)
```

Para la columna HORA ponemos esta media ya que tambien es nulo y creamos la funcion para cambiar de formato

```python
def convertir_hora(decimal_hora):
    # Extraer la parte entera y decimal de la hora
    horas_enteras = int(decimal_hora)
    minutos_decimales = (decimal_hora - horas_enteras) * 60

    # Formatear como cadena de tiempo (HH:MM)
    hora_formateada = f"{horas_enteras:02d}:{int(minutos_decimales):02d}"

    return hora_formateada
df_hechos_victimas['HORA'] = df_hechos_victimas['HORA'].replace('SD', convertir_hora(mediana))
df_hechos_victimas.head()
```

Verificamos ahora si aun tenemos valores nan

```python
# Verificar si hay NaN en la columna
hay_nan = df_hechos_victimas['HH'].isna().any()

# Mostrar el resultado
if hay_nan:
    print('La columna contiene valores NaN.')
else:
    print('La columna no contiene valores NaN.')
```

Creamos una funcion para que los valores sd cambien a nan

```python
def columns_modifier(DF,columns, value, replace=None):
    df=DF.copy()
    
    for index, row in df.iterrows():
        to_replace=row[columns[0]]

        if to_replace == value:
            df.at[index, columns[0]] = replace
    return df
```

convertimos algunas columnas numericas a nan los valores sd

```python
columna=['HORA']
df_hechos_victimas=columns_modifier(df_hechos_victimas, columna,'sd',np.nan)
```

```python
columna=['EDAD']
df_hechos_victimas=columns_modifier(df_hechos_victimas, columna,'SD',np.nan)
```

## Datos Adicionales Para los Graficos, analisis y kpi

### Columna semestre

Creamos una columna semestre que nos ayudar a una definicion de kpi

```python
# Crear la columna 'semestre'
df_hechos_victimas['semestre'] = df_hechos_victimas['FECHA_x'].dt.month // 7 + 1
df_hechos_victimas.head()
```

### Columna dia_semana_espanol

Crearemos la columna dia_semana_espanol para volver cada dia a texto perteneciente al nombre de la semana

```python
# Diccionario de traducción de días de la semana de inglés a español
dias_semana = {
    'Monday': 'Lunes',
    'Tuesday': 'Martes',
    'Wednesday': 'Miércoles',
    'Thursday': 'Jueves',
    'Friday': 'Viernes',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}

# Obtén el nombre del día de la semana y crea una nueva columna
df_hechos_victimas['nombre_dia'] = df_hechos_victimas['FECHA_x'].dt.strftime('%A')
df_hechos_victimas

# Change column type to string for column: 'nombre_dia'
df_hechos_victimas = df_hechos_victimas.astype({'nombre_dia': 'string'})
df_hechos_victimas['dia_semana_espanol'] = df_hechos_victimas['nombre_dia'].replace(dias_semana)
df_hechos_victimas.drop(columns=['nombre_dia'], inplace=True)
df_hechos_victimas.head()
```

### Columna Rango etario

Agrupamos la edad en rangos etarios para tener una mejor comprencion de los datos 'Infante', 'Niñez', 'Adolescencia', 'Adulto Joven', 'Adulto', 'Adulto de Edad Media', 'Adulto Mayor', 'Anciano'

```python
# Definir los límites de los rangos etarios
bins = [0, 5, 12, 18, 24, 35, 45, 60, 100]  # Puedes ajustar estos límites según tus necesidades

# Definir las etiquetas para cada rango etario
labels = ['Infante', 'Niñez', 'Adolescencia', 'Adulto Joven', 'Adulto', 'Adulto de Edad Media', 'Adulto Mayor', 'Anciano']

# Crear una nueva columna 'Rango Etario'
df_hechos_victimas['Rango Etario'] = pd.cut(df_hechos_victimas['EDAD'], bins=bins, labels=labels, right=False)
df_hechos_victimas.drop(columns=['EDAD'], inplace=True)
df_hechos_victimas.head()
```
