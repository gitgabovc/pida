```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)#permite ver todas las columnas de un df
```

```python
df_accidentes_homicidios = pd.read_csv("Dataset/df_hechos_victimas_clean_t.csv")
df_accidentes_homicidios.head()
```

```python
import folium
from folium.plugins import HeatMap

# Filtrar los puntos con valores de longitud y latitud diferentes de 0
df = df_accidentes_homicidios[(df_accidentes_homicidios['Latitud'] != 0) & (df_accidentes_homicidios['Longitud'] != 0)]

# Crear un mapa centrado en una ubicación específica
m = folium.Map(location=[-34.6037, -58.3816], zoom_start=12)

# Crear un mapa de calor
heat_data = [[row['Latitud'], row['Longitud']] for index, row in df.iterrows()]
HeatMap(heat_data, radius=15, gradient={0.2: 'blue', 0.4: 'green', 0.6: 'orange', 1: 'red'}).add_to(m)

# Guardar el mapa en un archivo HTML
m.save('mapa_personalizado.html')
```
